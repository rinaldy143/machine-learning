```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandasai import SmartDataframe
from pandasai.llm.openai import OpenAI



# Initialize PandasAI with your OpenAI key
key = ""  # Replace with your OpenAI API key
llm = OpenAI(api_token=key)
# pandas_ai = SmartDataframe(sales_by_country, config={"llm": llm})
# pandas_ai.chat('which country has the highest sales?')

```

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/16Gln4ZbzUA7z291LKxi-SEH-Q4mnZFC3R3jUyml1gho/edit#gid=1903509374'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
print(sheet_url_trf)

df = pd.read_csv(sheet_url_trf)
df.head()
```

```python
#checking info of table, from types, row, null etc
df.info()
```

```python
df.columns
```

```python
#change type date
df_cleaned = df.copy()

df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])
df_cleaned['Revenue'] = df_cleaned['Unit price'] * df_cleaned['Quantity']

print(df_cleaned.info())
df_cleaned.head()
```

```python
#check typo

df_cleaned['Gender'].value_counts()
```

```python
for col_name in ['Payment',  'Branch', 'City', 'Customer type',
       'Gender', 'Product line']:
  print(df_cleaned[col_name].value_counts(), '\n')
```

```python
#replace
dict_typo = {'male' : 'Male',
             'female':'Female',
             'fml': 'Female'}

df_replaced = df_cleaned.replace(dict_typo)
df_replaced.head()
```

```python
df_replaced['Gender'].value_counts()
```

```python
df_replaced['Invoice ID'].duplicated().sum()
```

# EDA

## daily trend overall

```python
agg_daily = df_replaced.groupby('Date', as_index=False)['Revenue', 'Quantity'].sum()
agg_daily.head()
```

```python
plt.rcParams["figure.figsize"] = (20,5)

plt.plot('Date','Revenue', data=agg_daily)
plt.title('Revenue trend')
plt.xlabel('Date')
plt.ylabel('Rev($)')
plt.show()
```

revenue of our data has seasonality

```python
sns.lineplot(data = agg_daily, x= 'Date', y='Quantity')
```

```python
pandas_ai = SmartDataframe(df_replaced, config={"llm": llm})
chart_ai = pandas_ai.chat('create trend line for revenue with seaborn with dark background')
chart_ai
```

```python
agg_payment = df_replaced.groupby('Payment', as_index=False).agg({'Revenue':'sum', 'Invoice ID':'nunique'})
agg_payment.head()
```

```python
plt.rcParams["figure.figsize"] = (10,5)
sns.barplot( x='Payment', y='Revenue', data = agg_payment.sort_values('Revenue', ascending=False))
```

```python
pandas_ai = SmartDataframe(df_replaced, config={"llm": llm})
chart_ai = pandas_ai.chat('create bar chart for each payment revenue with dark background')
chart_ai
```

highest revenue datang dari Cash

```python
sns.barplot(x='Payment', y='Invoice ID', data = agg_payment.sort_values('Invoice ID', ascending=False))
```

dari count transaction, ewallet paling banyak

```python
plt.pie(agg_payment['Revenue'], labels = agg_payment['Payment'],autopct='%.1f%%')
```

```python
pandas_ai = SmartDataframe(df_replaced, config={"llm": llm})
chart_ai = pandas_ai.chat('create pie chart of renevue each payment type with dark background')
chart_ai
```

cash volume 34.7% dan itu diatas 3.5% dari credit card

# By Gender and By Product Type

```python
agg_gend = df_replaced.groupby(['Gender', 'Product line'], as_index=False)['Invoice ID'].nunique()
agg_gend.head()
```

```python
plt.rcParams["figure.figsize"] = (15,5)
sns.barplot(x='Product line', y='Invoice ID', data = agg_gend, hue = 'Gender')
```
