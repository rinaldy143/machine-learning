```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "Sales_Data.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Untuk Melihat Baris dan Kolom data
df.info()
```

```python
# Untuk Melihat Nama Kolom
df.columns
```

```python
# Mengubah Date Menjadi Tanggal Universal dan menghapus data duplikat
df_cleaned = df.copy()
df_cleaned['ORDERDATE'] = pd.to_datetime(df_cleaned['ORDERDATE'])
print(df_cleaned.info())
df_cleaned.head()
```

```python
# Menghapus Typo
dict_typo = {'Atelier graphique' : 'Atelier Graphique'
             }

df_replaced = df_cleaned.replace(dict_typo)
df_replaced.head()
```

**Soal 1**

```python
# Filter hanya data dengan status "Shipped"
df_shipped = df[df['STATUS'] == 'Shipped']

# Agregasi jumlah produk yang terjual berdasarkan PRODUCTLINE
agg_product_sales = df_shipped.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().reset_index()

# Urutkan berdasarkan jumlah produk terjual
agg_product_sales = agg_product_sales.sort_values('QUANTITYORDERED', ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x='PRODUCTLINE', y='QUANTITYORDERED', data=agg_product_sales)

plt.title('Total Quantity Ordered per Product Line (Shipped Only)')
plt.xlabel('Product Line')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)
plt.show()
```

**Soal 2**

```python
# Filter hanya yang sudah dikirim
df_shipped = df[df['STATUS'] == 'Shipped']

# Pastikan ORDERDATE dalam format datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Buat total sales
df['TOTALSALES'] = df['PRICEEACH'] * df['QUANTITYORDERED']

# Resample per bulan agar lebih rapi
agg_monthly_sales = df.resample('M', on='ORDERDATE')['TOTALSALES'].sum()

# Plot line chart dengan rentang waktu per bulan
plt.figure(figsize=(15, 5))
plt.plot(agg_monthly_sales.index, agg_monthly_sales.values, marker='o', linestyle='-')

plt.title('Sales Performance Over Time (Monthly)')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

**Soal 3**

```python
# Filter hanya data dengan status "Shipped"
df_shipped = df[df['STATUS'] == 'Shipped']

# Agregasi jumlah produk terjual berdasarkan DEALSIZE
agg_dealsize = df_shipped.groupby('DEALSIZE')['QUANTITYORDERED'].sum().reset_index()

# Plot pie chart
plt.figure(figsize=(7, 7))
plt.pie(agg_dealsize['QUANTITYORDERED'], labels=agg_dealsize['DEALSIZE'], autopct='%.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])

plt.title('Deal Size vs Quantity Ordered (Shipped Orders Only)')
plt.show()
```
